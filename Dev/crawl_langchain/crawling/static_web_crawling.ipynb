## requests

```python
# requests 패키지 가져오기
import requests               

# 가져올 url 문자열로 입력
url = 'https://www.naver.com/'  
# requests의 get함수를 이용해 해당 url로 부터 html이 담긴 자료를 받아옴
response = requests.get(url)

# 우리가 얻고자 하는 html 문서가 여기에 담기게 됨
html_text = response.text
print(html_text)
```

```python
from lxml.html import fromstring
xpath = '/html/body'
data = fromstring(html_text).xpath(xpath)
for a in data:
    print(a.text_content())
```

```python
# BeautifulSoup 패키지 불러오기
# 주로 bs로 이름을 간단히 만들어서 사용함
from bs4 import BeautifulSoup as bs

# html을 잘 정리된 형태로 변환
html = bs(html_text, 'html.parser')
html
```

## find/select

### find( ), find_all( ) 함수

```python
# 목표 태그 예)
'''
<p class = "para">코딩유치원</p>
<div id = "zara">코딩유치원</p>
'''

# 태그 이름으로 찾기
p_tag = html.find('p')

# 태그 속성(class)으로 찾기 - 2가지 형식
para_calss = html.find(class_='dsc') # 이 형식을 사용할 때는 class 다음에 언더바_를 꼭 붙여주어야 한다
para_attrs_calss = html.find(attrs = {'class':'dsc'}) 

# 태그 속성(id)으로 찾기 - 2가지 형식
zara_id = html.find(id='wrap') 
zara_attrs_id = html.find(attrs = {'id':'wrap'})

# 태그 이름과 속성으로 찾기
p_tag_para_calss = html.find('p', class_='dsc')
div_tag_zara_id = html.find('div', {'id': 'wrap'})
```

```python
print("p_tag: ", p_tag)
print("para_calss == para_attrs_calss:", para_calss == para_attrs_calss)
print("zara_id == zara_attrs_id:", zara_id == zara_attrs_id)
print("p_tag_para_calss: ", p_tag_para_calss)
print("div_tag_zara_id: ", div_tag_zara_id)
```

### select( ), select_one( ) 함수

```python
# a태그의 class 속성명이 news_tit인 태그 
title = html.select_one('a.link_dsc')
titles = html.select('a.link_dsc')

print("link_dsc:", title.get_text())
print("link_dsc:\n")
for i in titles: 
    title = i.get_text()
    print(title)
```

# 정적 웹크롤링 - 텍스트, 하이퍼링크, 이미지

```python
# 크롤링 할 페이지(url) HTML 가져오기

# step1.프로젝트에 필요한 패키지 불러오기
from bs4 import BeautifulSoup as bs
import requests

# step2. 검색할 키워드 입력
query = input('검색할 키워드를 입력하세요: ')

# step3. 입력받은 query가 포함된 url 주소(네이버 뉴스 검색 결과 페이지) 저장
url = 'https://search.naver.com/search.naver?where=news&sm=tab_jum&query='+'%s'%query

# step4. requests 패키지를 이용해 'url'의 html 문서 가져오기
response = requests.get(url)
html_text = response.text

# step5. beautifulsoup 패키지로 파싱 후, 'soup' 변수에 저장
soup = bs(response.text, 'html.parser')
```

```python
# step6.뉴스 제목 텍스트 추출 / .get_text( )
news_titles = soup.select("a.news_tit")

print("News Titles:")
for i in news_titles:
    title = i.get_text()
    print(title)
```

```python
# step7.뉴스 하이퍼링크 추출 / .attrs['href']
print("News Links:")
for i in news_titles:
    href = i.attrs['href']
    print(href)
```

```python
# from IPython.display import Image
# Image(filename='test.png')
# thum_xpath = '//*[@id="sp_nws1"]/div[1]/div/div[2]/a[1]/img'
thum_xpath = '//*[contains(@id, "sp_nws")]/div[1]/div/div[2]/a[1]/img'
thum_imgs = fromstring(requests.get('https://search.naver.com/search.naver?where=news&sm=tab_jum&query=%EA%B1%B4%EA%B0%95').text).xpath(thum_xpath)
# Image(thum_img)
# print(dir(thum_img))
for i, thum_img in enumerate(thum_imgs):
    with open(f'./crawling/img_download/test{i}.png', 'wb') as jpg:
        jpg.write(requests.get(thum_img.attrib['data-lazysrc']).content)
```

```python
#step8.뉴스 썸네일 이미지 추출 / .attrs['src']
news_content_div = soup.select(".news_contents")
print(news_content_div)
news_thumbnail = [thumbnail.select_one(".thumb") for thumbnail in news_content_div]
link_thumbnail = []

for img in news_thumbnail:
    if img is not None and 'data-lazysrc' in img.attrs:
        link_thumbnail.append(img.attrs['data-lazysrc'])

# 이미지 저장할 폴더 생성
import os

# path_folder의 경로는 각자 저장할 폴더의 경로를 적어줄 것
path_folder = './crawling/img_download/'

if not os.path.isdir(path_folder):
    os.mkdir(path_folder)

# 이미지 다운로드
from urllib.request import urlretrieve

i = 0

for link in link_thumbnail:          
    i += 1
    urlretrieve(link, path_folder + f'{i}.jpg')
```
